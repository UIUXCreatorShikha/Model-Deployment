```python
import pandas as pd
```

```python
dft=pd.read_csv('tested.csv')

```

```python
dft
```

```python
dft.isnull().sum()
```

```python
dft['Age'] = dft['Age'].fillna(dft['Age'].mean())
dft['Cabin'] = dft['Cabin'].fillna('Unknown')
dft['Embarked'] = dft['Embarked'].fillna(dft['Embarked'].mode()[0])
dft['Fare'] = dft['Fare'].fillna(dft['Fare'].mean())
```

```python
dft
```

```python
dft = dft.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
```

```python
dft['Sex'] = dft['Sex'].map({'male': 0, 'female': 1})
dft = pd.get_dummies(dft, columns=['Embarked'], drop_first=True)
```
